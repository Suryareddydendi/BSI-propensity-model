```python
%%time
# Transformation Mck Oneoff dataset (Reamaining Term, Bucketing FICO Score, Current LTV)
mck_oneoff_specific_column["Interest_Rate_Date"]=pd.to_datetime(mck_oneoff_specific_column["Interest_Rate_Date"], format='%Y-%m')
mck_oneoff_specific_column['Year'] = mck_oneoff_specific_column['Interest_Rate_Date'].dt.strftime('%Y')
mck_oneoff_specific_column['Month'] = mck_oneoff_specific_column['Interest_Rate_Date'].dt.strftime('%m')
mck_oneoff_specific_column['Term'] = np.where((mck_oneoff_specific_column['RemainingTerm']/12) > 18,30,15) # 30, 15 years 
 
#Bucketing FICO SCORE
col = 'FICO_Score'
conditions = [
    mck_oneoff_specific_column[col].between(0,620),   # inclusive = True is the default
    mck_oneoff_specific_column[col].between(620,640),
    mck_oneoff_specific_column[col].between(640,660),
    mck_oneoff_specific_column[col].between(660,680),
    mck_oneoff_specific_column[col].between(680,700),
    mck_oneoff_specific_column[col].between(700,720),
    mck_oneoff_specific_column[col].between(720,740),
    mck_oneoff_specific_column[col].between(740,760),
    mck_oneoff_specific_column[col].between(760,780),
    mck_oneoff_specific_column[col].ge(780)
             ]
choices = ["<620",">=620",">=640",">=660",">=680",">=700",">=720",">=740",">=760",">=780"]
mck_oneoff_specific_column["FICO"] = np.select(conditions, choices, default=np.nan) # If not following this conditions any data point by default put NaN values.
 
#Bucketing LTV
col = 'CurrentLTV'
conditions = [
    mck_oneoff_specific_column[col].between(0,30),   # inclusive = True is the default
    mck_oneoff_specific_column[col].between(30,60), 
    mck_oneoff_specific_column[col].between(60,70),
    mck_oneoff_specific_column[col].between(70,75),
    mck_oneoff_specific_column[col].between(75,80),
    mck_oneoff_specific_column[col].between(80,85),
    mck_oneoff_specific_column[col].between(85,90),
    mck_oneoff_specific_column[col].between(90,95),
    mck_oneoff_specific_column[col].ge(95)
             ]
choices = ["<=30","<=60","<=70","<=75","<=80","<=85","<=90","<=95",">95"]
mck_oneoff_specific_column["LTV"] = np.select(conditions, choices, default=np.nan) # If not following this conditions any data point by default put NaN values.
mck_oneoff_specific_column.head(10)
mck_oneoff_specific_column.shape
```
